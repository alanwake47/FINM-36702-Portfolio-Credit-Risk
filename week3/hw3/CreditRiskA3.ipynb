# Question 1

Let ùë¶ = ùëôùëîùëë[ùëëùëü], <br>
then ùëëùëü = ùë¶^2, <br><br> and we have:<br>
ùëùùëëùëì_ùëôùëîùëë[ùëôùëîùëë] = ùëùùëëùëì_ùëëùëü[ùëëùëü] ùëë(ùëëùëü)/ùëë(ùëôùëîùëë) = ùëùùëëùëì_ùëëùëü[ùë¶^2] 2ùë¶
<br><br>
Substituting the expression for ùëùùëëùëì_ùëëùëü[ùëëùëü], we get:
ùëùùëëùëì_ùëôùëîùëë[ùëôùëîùëë] = (2 - 2ùë¶^2)2ùë¶ = 4ùë¶ - 4ùë¶^3

```python
def default_rate_pdf(dr):
    return 2 - 2 * dr

def lgd_pdf(lgd):
    return (4*lgd)-(4*lgd**3)
```

```python
x = np.linspace(0, 1, 100)
plt.plot(x, default_rate_pdf(x), label='Default Rate')
plt.plot(np.sqrt(x), lgd_pdf(np.sqrt(x)), label='LGD')

plt.xlabel('Variable')
plt.ylabel('Probability Density Function')
plt.legend()

plt.show()
```
